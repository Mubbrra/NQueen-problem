```python
from copy import deepcopy                        
import random                                  
from math import exp                            
import numpy as np


def placeQueens(board):                          
    i = 0
    while (i < 8):
        row = random.randint(0,8 - 1)
        if (board[row][i] != "Q"):
            board[row][i] = "Q"
            i+=1

def getQueens(board):                            # This will get the positions of queens palced in the maze
    queen_positions = []                         
    for i in range(8):
        for j in range (8):
            if board[i][j] == "Q":
                temp = i,j
                queen_positions.append(temp)
    return queen_positions

def printBoard(board):                          
    for i in range(8):
        for j in range(8):
            print(board[i][j], end=' ')
        print()
def getBoard(board,queen_pos):
    boards = []
    for i in range(len(queen_pos)):
        x = deepcopy(queen_pos[i][0])
        y = deepcopy(queen_pos[i][1])
        for j in range(8):
            if(x!=j):
                b = deepcopy(board)
                b[x][y] = "*"
                b[j][y] = "Q"
                boards.append(b)
    return boards
def objective_function(board):    

    queenattack=0
    
    cqueen = 0
    
   
    # queens in diagonal
    
    
    for i in range(8):
        cqueen = 0
        for j in range (8):
            cqueen = 0
            if board[i][j] == "Q":
              
            
            
                x = deepcopy(i)
                y = deepcopy(j)
                while(x>=0 & y >=0):
                    if board[x][y] == "Q":
                        cqueen=cqueen+1
                        if cqueen>1:
                            queenattack=queenattack+1
                    x-=1
                    y-=1
                    
               
            
            
                x = deepcopy(i)
                y = deepcopy(j)
                cqueen = 0
                while(x>=0 & y<8):
                    if(y>=8):
                        break
                    if board[x][y] == "Q":
                        cqueen=cqueen+1
                        if cqueen>1:
                            queenattack=queenattack+1
                    x-=1
                    y+=1
                    
               
            
                x = deepcopy(i)
                y = deepcopy(j)
                cqueen = 0
                while(x<8 & y <8):
                    if board[x][y] == "Q":
                        cqueen=cqueen+1
                        if cqueen>1:
                            queenattack=queenattack+1
                    x+=1
                    y+=1
                    
                    
               
                
                
                x = deepcopy(i)
                y = deepcopy(j)
                cqueen = 0
                while(x<8 & y >=0):
                    if board[x][y] == "Q":
                        cqueen=cqueen+1
                        if cqueent>1:
                            queenattack=queenattack+1
                    x+=1
                    y-=1
               
                    
     #queens in same row
    for i in range(8):
        cqueen = 0
        for j in range (8):
            if board[i][j] == "Q":
                cqueen=cqueen+1
                if cqueen>1:
                    queenattack=queenattack+1
                    
                    
    return queenattack
```

```python
def hillClimbing(board): 
    
    queen_position = getQueens(board)
    
    heu = objective_function(board)
    
    board_list= getBoard(board,queen_position)
    harray = []
    
    for i in board_list:
        
        harray.append(objective_function(i))
    min_h = board_list[harray.index(min(harray))] 
    if min(harray) == 0:
        
        return min_h
    if min(harray) < heu:
        
        return min_h 
    return hillClimbing(min_h)
   
def main():
    
   
    
    rows , cols = (8,8)
    board = []                                       

    for i in range(rows):                         
        col = []
        for j in range(cols):                       
            col.append("*")
        board.append(col)
    
    placeQueens(board)
    printBoard(board)

    print("QUEENS ATTACKING: ",objective_function(board))
    final_board = hillClimbing(board) 
    print("FINAL CONFIGURATION OF BOARD: ")
    printBoard(final_board)
    
if __name__ == "__main__":
    main()
```
